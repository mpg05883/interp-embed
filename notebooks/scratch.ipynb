Load the [GSM8k dataset](https://huggingface.co/datasets/openai/gsm8k) from Hugging Face as a pandas DataFrame.

```python
import pandas as pd
from datasets import load_dataset

dataset = load_dataset(path="gsm8k", name="main")

df = pd.concat(
    [
        dataset["train"].to_pandas().assign(split="train"),
        dataset["test"].to_pandas().assign(split="test"),
    ],
    ignore_index=True,
).reset_index(names="idx")

print(f"Number of rows: {len(df)}")
df.head()
```

```python
type(df["question"])
```

View the train split.

```python
df_train = df[df["split"] == "train"]

print(f"Number of rows: {len(df_train)}")
df_train.head()
```

View the test split.

```python
df_test = df[df["split"] == "test"]

print(f"Number of rows: {len(df_test)}")
df_test.head()
```
